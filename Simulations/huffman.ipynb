```python
import numpy as np
%matplotlib qt
import matplotlib.pyplot as plt
```

```python
class node:
    def __init__(self,nodelabel,nodeprob):
        self.label = nodelabel
        self.prob = nodeprob
        self.leftChild = None 
        self.rightChild = None
```

```python
node1 = node('A',0.5)
```

```python
print(node1.label,node1.prob,node1.leftChild,node1.rightChild)
```

```python
node2 = node('B',0.25)
node3 = node('C',0.25)
node1.leftChild = node2
node1.rightChild = node3 
print(node1.label, node1.prob, node1.leftChild, node1.rightChild)
```

```python
print(node1.leftChild.label, node1.rightChild.label)
```

```python
def merge_nodes(node1,node2):
    mergelabel = node1.label+node2.label 
    mergeprob = node1.prob+node2.prob 
    mergenode = node(mergelabel,mergeprob)
    mergenode.leftChild = node1
    mergenode.rightChild = node2
    return mergenode
```

```python
node4 = merge_nodes(node1,node2)
print(node4.label,node4.prob,node4.leftChild.label,node4.rightChild.label)
```

```python
alphabet = ['A','B','C','D','E']
p = 1/(np.arange(len(alphabet))+1)
p = p/np.sum(p)
print(p)
```

```python
def create_huffman_tree(alphabet,p):
    nodelist = []
    for i in range(len(alphabet)):
        nodelist.append(node(alphabet[i],p[i]))
    print([n.label for n in nodelist])
    while(len(nodelist)>1):
        sort_order = np.argsort([n.prob for n in nodelist])
        node1 = nodelist[sort_order[0]]
        node2 = nodelist[sort_order[1]]
        merged_node = merge_nodes(node1,node2)
        print('Remove nodes ',node1.label, ' and ',node2.label, ' and add ',merged_node.label)
        nodelist.remove(node1)
        nodelist.remove(node2)
        nodelist.append(merged_node)
        print([n.label for n in nodelist])
        print([n.prob for n in nodelist])
    return nodelist[0]
```

```python
rootnode = create_huffman_tree(alphabet,p)
```

```python
def encode_huffman(rootnode,s):
    codeword = ''
    currentnode = rootnode 
    while currentnode.leftChild is not None and currentnode.rightChild is not None:
        if s in currentnode.leftChild.label:
            codeword += '0'
            currentnode = currentnode.leftChild 
        else:
            codeword += '1'
            currentnode = currentnode.rightChild
    return codeword
```

```python
for s in alphabet:
    print('Codeword for ',s,' is ',encode_huffman(rootnode,s))
```

```python
def decode_huffman(rootnode,rec_string):
    decodedstring = ''
    currentnode = rootnode  
    for l in range(len(rec_string)):
        if rec_string[l] == '0':
            currentnode = currentnode.leftChild 
        else:
            currentnode = currentnode.rightChild
        if currentnode.leftChild is None and currentnode.rightChild is None:
            decodedstring += currentnode.label
            currentnode = rootnode 
    return(decodedstring)
```

```python
for s in alphabet:
    c = encode_huffman(rootnode,s)
    d = decode_huffman(rootnode,c)
    print('Symbol ',s,' encoded to ',c,', decoded as ',d)
```

```python
N = 1000
rand_symbols = np.random.choice(len(alphabet),N,p=p)
randomstring = ''
print(rand_symbols)
for i in rand_symbols:
    randomstring += alphabet[i]
print(randomstring)
plt.hist(rand_symbols)
plt.show()
```

```python
encodedstring = ''
for s in randomstring:
    encodedstring += encode_huffman(rootnode,s)
print(encodedstring)
```

```python
decodedstring = decode_huffman(rootnode,encodedstring)
if decodedstring == randomstring:
    print('Decoded correctly')
else:
    print('Decoded incorrectly')
```

```python
print('Rate of the compressed string: ',len(encodedstring)/N)
freq = np.zeros(len(alphabet))
for i in range(len(alphabet)):
    for s in randomstring:
        if s==alphabet[i]:
            freq[i] += 1
print(freq/N)
print('Entropy: ',np.sum(p*np.log2(1/p)))
```

```python
allowedsym = 'abcdefghijklmnopqrstuvwxyz'
char_count = {c:0 for c in allowedsym}
randomstring = ''
with open('macbeth.txt','r') as f:
    text = f.read()
    for c in text:
        if c in allowedsym:
            char_count[c] += 1
            randomstring += c
print(len(randomstring))
print(char_count)
```

```python
p = np.array(list(char_count.values()))
p = p/np.sum(p)
print(p,np.size(p))
```

```python
rootnode = create_huffman_tree(list(char_count.keys()),p)
```

```python
encodedstring = ''
for s in randomstring:
    encodedstring += encode_huffman(rootnode,s)
decodedstring = decode_huffman(rootnode,encodedstring)
if decodedstring == randomstring:
    print('Decoded correctly')
else:
    print('Incorrect decoding')
```
